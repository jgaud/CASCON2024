```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.gridspec as gridspec
```

```python
xls = pd.ExcelFile('results.xlsx')
cols = ['Type', 'Baseline', 'No Threshold', 'With Threshold']
sheet_names = ['NSLKDD', 'NDSec1', 'CIC IDS 2017', 'NF-UNSW-NB15-v2', 'NF-CSE-CIC-IDS2018-v2']
dfs = []

for name in sheet_names:
    df = pd.read_excel(xls, name, names=cols)
    df.set_index('Type')
    dfs.append(df)
```
